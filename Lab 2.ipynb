```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Генерація даних
X = np.arange(0, 10, 0.1)
Y_true = 2 * X + 1
Y_noise = Y_true + np.random.normal(0, 1, len(X))  # Додавання шуму

# Обчислення метрик похибки
mae = mean_absolute_error(Y_true, Y_noise)
mse = mean_squared_error(Y_true, Y_noise)

# Побудова графіка
plt.scatter(X, Y_noise, label='Згенеровані дані з шумом')
plt.plot(X, Y_true, color='red', label='Теоретична лінія')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Запис результатів у CSV файл
import csv
with open('results.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["X", "Y_true", "Y_noise", "mAE", "mSE"])
    for x, y_true, y_noise in zip(X, Y_true, Y_noise):
        writer.writerow([x, y_true, y_noise, mae, mse])
```
