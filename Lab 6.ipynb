```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

#  ,   
scalar = tf.constant(3.0)
vector = tf.constant([1.0, 2.0, 3.0])
matrix = tf.constant([[1.0, 2.0], [3.0, 4.0]])

print("Scalar:", scalar)
print("Vector:", vector)
print("Matrix:", matrix)

# Create -  
tensor = tf.constant([[5, 7], [9, 11]])

# Read -  
print("Tensor:", tensor)

# Update -    (  )
updated_tensor = tf.add(tensor, 5)
print("Updated Tensor:", updated_tensor)

# Delete -  (TensorFlow    ,    )
zeroed_tensor = tf.zeros_like(tensor)
print("Zeroed Tensor:", zeroed_tensor)

#     
X_reg = np.random.rand(100, 1)
y_reg = 3 * X_reg + 2 + np.random.randn(100, 1)

# 
plt.scatter(X_reg, y_reg, color="blue")
plt.title("   ")
plt.xlabel("X")
plt.ylabel("y")
plt.show()

#        
X_clf, y_clf = make_classification(n_samples=100, n_features=2, n_classes=2, n_clusters_per_class=1,
                                   n_informative=2, n_redundant=0, n_repeated=0)

# 
plt.scatter(X_clf[:, 0], X_clf[:, 1], c=y_clf, cmap="viridis")
plt.title("   ")
plt.xlabel(" 1")
plt.ylabel(" 2")
plt.show()
```
