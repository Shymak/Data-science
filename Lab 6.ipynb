```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# Створення скаляра, вектора і матриці
scalar = tf.constant(3.0)
vector = tf.constant([1.0, 2.0, 3.0])
matrix = tf.constant([[1.0, 2.0], [3.0, 4.0]])

print("Scalar:", scalar)
print("Vector:", vector)
print("Matrix:", matrix)

# Create - Створення тензорів
tensor = tf.constant([[5, 7], [9, 11]])

# Read - Читання даних
print("Tensor:", tensor)

# Update - Оновлення значень тензора (створення нового тензора)
updated_tensor = tf.add(tensor, 5)
print("Updated Tensor:", updated_tensor)

# Delete - Видалення (TensorFlow не підтримує пряме видалення, але можна замінити значення)
zeroed_tensor = tf.zeros_like(tensor)
print("Zeroed Tensor:", zeroed_tensor)

# Генерація випадкових даних для регресії
X_reg = np.random.rand(100, 1)
y_reg = 3 * X_reg + 2 + np.random.randn(100, 1)

# Візуалізація
plt.scatter(X_reg, y_reg, color="blue")
plt.title("Генеровані дані для регресії")
plt.xlabel("X")
plt.ylabel("y")
plt.show()

# Генерація випадкових даних для класифікації з коректними параметрами
X_clf, y_clf = make_classification(n_samples=100, n_features=2, n_classes=2, n_clusters_per_class=1,
                                   n_informative=2, n_redundant=0, n_repeated=0)

# Візуалізація
plt.scatter(X_clf[:, 0], X_clf[:, 1], c=y_clf, cmap="viridis")
plt.title("Генеровані дані для класифікації")
plt.xlabel("Фіча 1")
plt.ylabel("Фіча 2")
plt.show()
```
